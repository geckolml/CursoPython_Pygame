# Travesia
***

### Descripcion

* Travesia es un juego para 2 jugadores, el cual consiste principalmente en ir uniendo puntos del color que le corresponda a cada jugador hasta formar un camino desde un lado del tablero hasta el otro lado contrario sin chocar las lineas del otro jugador.

### Cantidad de jugadores 

* 2 jugadores(rojo y azul)

### Tipo de juego

* Es un juego de estrategia por turnos.

### Nota
* El jugador ***rojo*** siempre empieza.
* Hacer click en 2 puntos contiguos del color que le corresponda para hacer una jugada.

### Codigo fuente

```python
# coding=utf-8
########################################
## Desarrollador : Renzo Elvis Cerron Tome
## Computer Science - UNI
## Taller de Python - ACECOM - Verano 2017
########################################

## Game Bridg-It

import pygame as pg # importamos pygame
from pygame.locals import *
import numpy as np

class Line(pg.sprite.Sprite):
	def __init__(self,color,width,height):
		super.__init__()
		self.image = pg.Surface([width,height])
		self.image.fill(color)

		self.rect = self.image.get_rect()


pg.init() #Inicializamos pygame
BLACK = (0,0,0)      # Color negro
BLUE = (0,0,155)     # Color azul
RED = (155,0,0)      # Color rojo
YELLOW = (155,155,0) # Color amarillo
size = (800,700)    # TamaÃ±o de la pantalla

bluePoint=pg.image.load("images/bluePoint.png") # Puntos azules que son partes del tablero de juego
redPoint=pg.image.load("images/redPoint.png")   # Puntos rojos que son partes del tablero de juego
tablero = pg.image.load("images/table.png")     # Tablero del juego 


screen = pg.display.set_mode(size) # Se crea objeto Surface 
pg.display.set_caption("Bridg-It") # Se asigna el titulo

done = False # Variable bandera que indica si el programa acaba o no acaba, False : Todavia no acaba
dy=100 # Distancia en el eje x de inicio del tablero
dx=40  # Distancia en el eje y de inicio del tablero
xr=70  # Distancia entre puntos
color = BLUE # Nos indica el color del turno actual .
cont=0 # Variable Contador que nos ayudara a determinar de quien es el turno. 
       # cont par : Rojo, cont impar:Azul
coord=[] # Variable de almacenamiento de las coordenadas para analizar si es un movimiento valido o no
moves_blue=[] # Se guardan las coordenadas de los puntos que se unen para formar la linea que eligio el jugador azul
moves_red=[] # Se guardan las coordenadas de los puntos que se unen para formar la linea que eligio el jugador rojo
x,y=0,0 # Nos indica la posicion del mouse donde hace click
jugada_valida = False # Nos indica si la jugada fue valida o no.

ganador = 0 # 0 : No ganador, 1: Gana Rojo, 2 : Gana Azul


msj = "" # Mensaje que indica quien gano

R = np.zeros((56,56))
B = np.zeros((56,56))

visitados1 = np.zeros(56) # Para el DFS Rojo
visitados2 = np.zeros(56) # Para el DFS Azul

lineas_azules = pg.sprite.Group()

lineas_rojas  = pg.sprite.Group()

todas_lineas  = pg.sprite.Group()

ix1,jy1,ix2,jy2 = 0, 0, 0, 0

##################
#        DFS     #
##################

def dfs(arr,vis,root):
    vis[root] = 1
    for i in range(56):
        if arr[root][i]!=0 and vis[i] == 0 :
            dfs(arr,vis,i)

#############################            
## Check - Revisar si gana ##
#############################

# ROJO
def check1(arr,vis,candidatos): # Chequear si jugador Rojo gana| de izquierda a derecha | 7 x 8 | 
    global ganador
    for s in candidatos: # Los candidatos son los puntos Rojos de la derecha que fueron seleccionados
        dfs(arr,vis,s)
        if sum(vis[-8:])>0:
            ganador = 1 # Gana jugador Rojo
            #print("Gano Jugador Rojo")
            return
        for i in range(len(vis)): # Setea a todos igual a cero
                vis[i]=0

#AZUL  
def check2(arr,vis,candidatos): # Chequear si jugador Azul gana | de izquierda a derecha | 7 x 8
    global ganador
    for s in candidatos: # # Los candidatos son los puntos Azules de la parte superior que fueron seleccionados
        dfs(arr,vis,s)
        if sum([vis[7],vis[15],vis[23],vis[31],vis[39],vis[47],vis[55]])>0:
            ganador = 2 # Gana jugador Azul
            #print("Gano Jugador Azul")
            return
        for i in range(len(vis)): # Setea a todos igual a cero
                vis[i]=0  

######################
##    MAPEO         ##
######################

# Mapea coordenadas c(x,y) en indices de matriz i,j respecto 
# a las posiciones del tablero
def mapeo(turno,c):
    i = 0
    j = 0
    if turno == 0: # Turno jugador Rojo, seteamos su matriz
        i=((c[1] - 170 )/35)/2
        j=((c[0] - 40 )/35)/2
        #R[i,j]=1
    else: # Turno jugador Azul, seteamos su matriz
        i=((c[1] - 135 )/35)/2
        j=((c[0] - 75 )/35)/2
        #B[i,j]=1
 	
    return i, j
    

while not done:
    for event in pg.event.get(): # Atento a los eventos de Salida, Teclado y Mouse
        if event.type == pg.QUIT:
            done = True
        elif event.type == pg.MOUSEBUTTONDOWN:
            #if pg.mouse.get_pressed()[0]:
                x,y= event.pos#pg.mouse.get_pos()
                coord.append((x,y))
        elif event.type == pg.KEYDOWN: # Si el usuario presiona una tecla
            if event.key == pg.K_ESCAPE:
                done = True
   
    screen.fill(BLACK) # Setea el background como negro
   
    screen.blit(tablero,[xr/2+dx-25,xr/2+dy-30,8*xr,8*xr]) # Se dibuja el tablero
    # Se dibujan las imagenes de los puntos azules encima del tablero de juego
    for i in range(7):
        for j in range(8):
            screen.blit(bluePoint,[xr+xr*i+dx,xr/2+xr*j+dy])
    # Se dibujan las imagenes de los puntos rojos encima del tablero de juego
    for i in range(8):
        for j in range(7):
            screen.blit(redPoint, [xr/2+xr*i+dx,xr+xr*j+dy])
    #_______________________________________________________    
    # Analizamos las jugadas en los turnos de los jugadores
    #--------------------------------------------------------
    if len(coord) == 2 : # Se hizo una jugada
        # Se analiza la validez de la jugada
        jugada_valida = Rect([xr/2+dx-25,xr/2+dy-30,8*xr,8*xr]).collidepoint(coord[0]) and Rect([xr/2+dx-25,xr/2+dy-30,8*xr,8*xr]).collidepoint(coord[1])
        if jugada_valida:
            #print("Movimiento Correcto | Rect= {}",tablero.get_rect())
            #print("{}  {}".format(coord[0],coord[1]))
            if cont % 2 == 0: # Empieza a jugar el rojo
                ix1,jy1 = mapeo(0,coord[0]) # Mapeo Matriz roja coordenada 0
                ix2,jy2 = mapeo(0,coord[1]) # Mapeo Matriz roja coordenada 1 
                moves_red.append([jy1*70 + 85,ix1*70+180])
                moves_red.append([jy2*70 + 85,ix2*70+180])
                R[ix1+7*jy1][ix2+7*jy2]=1
                R[ix2+7*jy2][ix1+7*jy1]=1
                check1(R,visitados1,range(7)) # Chequea de izquierda a derecha
            else:
                
                ix1,jy1 =mapeo(1,coord[0]) # Mapeo Matriz Azul coordenada 0
                ix2,jy2 = mapeo(1,coord[1]) # Mapeo Matriz Azul coordenada 1
                moves_blue.append([jy1*70 + 120,ix1*70+145])
                moves_blue.append([jy2*70 + 120,ix2*70+145]) 
                B[ix1+8*jy1][ix2+8*jy2] = 1
                B[ix2+8*jy2][ix1+8*jy1] = 1
                check2 (B,visitados2,[0,8,16,24,32,40,48]) # Chequea de arriba hacia abajo 
            cont=cont+1 # Despues de cada jugada valida aumenta en 1 para pasarle el turno al otro jugador

        del(coord[1]) # Eliminara la jugada si es o no es valida
        del(coord[0])
        #print(moves_blue)
        #print(moves_red)
              
    
    for i in range(0,len(moves_blue),2): # Dibuja todas las jugadas del azul
        pg.draw.line(screen,BLUE,moves_blue[i],moves_blue[i+1],3)  
        #
        #pg.draw.line(screen,BLUE,[jy1*70+135,ix1*70 + 75],[jy2*70+135,ix2*70 + 75],3)
    
    for i in range(0,len(moves_red),2):# Dibuja todas las jugadas del rojo
        #pg.draw.line(screen,RED,[jy1*70+170,ix1*70 + 40],[jy2*70+170,ix2*70 + 40],3)
		pg.draw.line(screen,RED,moves_red[i],moves_red[i+1],3)
        

    #todas_lineas.draw(screen)
    font_renderer = pg.font.Font('freesansbold.ttf',30)
    label = font_renderer.render(msj, 1,(255,255,255)) # RGB Color
    screen.blit(label,(200,50)) 
    if ganador == 1:
        msj="Gana Jugador Rojo"
        
    if ganador == 2:
        msj="Gana Jugador Azul"      
    
    pg.display.update()
pg.quit()
```
